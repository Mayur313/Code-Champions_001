```python
import numpy as np
import pandas as pd
```

```python
orders=pd.read_csv("olist_orders_dataset.csv")
```

```python
payments=pd.read_csv("olist_order_payments_dataset.csv")
```

```python
merge_order_payment = pd.merge(orders, payments, on='order_id', how='inner')
```

```python
merge_order_payment
```

```python
merge_order_payment.isnull().sum()
```

```python
corr_inst_val = merge_order_payment[['payment_installments', 'payment_value']].corr()
```

```python
corr_inst_val
```

```python
items=pd.read_csv("olist_order_items_dataset.csv")
```

```python
merge_order_payment = pd.merge(merge_order_payment, items, on='order_id', how='inner')
```

```python
merge_order_payment
```

```python
merge_order_payment['order_purchase_timestamp'] = pd.to_datetime(merge_order_payment['order_purchase_timestamp'])
merge_order_payment['year'] = merge_order_payment['order_purchase_timestamp'].dt.year
merge_order_payment['month'] = merge_order_payment['order_purchase_timestamp'].dt.month
order_yr_mn = merge_order_payment[['order_id', 'year', 'month']]

print(order_yr_mn)
```

```python
order_yr_mn["year"].max()
```

```python
df_2022 = order_yr_mn[order_yr_mn['year'] == 2017]
monthly_orders = df_2022.groupby('month')['order_id'].count()
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(monthly_orders.index, monthly_orders.values)
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Order Trends in 2017')
plt.show()
```
