```python
merge_order_payment['order_purchase_timestamp'] = pd.to_datetime(merge_order_payment['order_purchase_timestamp'])
merge_order_payment['year'] = merge_order_payment['order_purchase_timestamp'].dt.year
merge_order_payment['month'] = merge_order_payment['order_purchase_timestamp'].dt.month
order_yr_mn = merge_order_payment[['order_id', 'year', 'month']]

print(order_yr_mn)
```

```python
order_yr_mn["year"].max()
```

```python
df_2022 = order_yr_mn[order_yr_mn['year'] == 2017]
monthly_orders = df_2022.groupby('month')['order_id'].count()
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(monthly_orders.index, monthly_orders.values)
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Order Trends in 2017')
plt.show()
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'merge_order_payment'
# with columns: customer_id, customer_city, customer_unique_id, order_id

# Calculate total spending per customer
total_spending = merge_order_payment.groupby('customer_unique_id')['order_id'].count()

# Calculate purchase frequency per customer
purchase_frequency = merge_order_payment.groupby('customer_unique_id')['order_id'].nunique()

# Create a new DataFrame for segmentation
customer_segmentation = pd.DataFrame({
    'TotalSpending': total_spending,
    'PurchaseFrequency': purchase_frequency
})
threshold_value=5
# Now you can define segments based on these metrics
# For example, high-value customers:
high_value_customers = customer_segmentation[customer_segmentation['TotalSpending'] > threshold_value]

# Dormant customers (assuming threshold for inactivity is X months):
dormant_customers = customer_segmentation[customer_segmentation['PurchaseFrequency']  <2]

# You can adjust the threshold values based on your business context.
```

```python


# Assuming you have a DataFrame named 'merge_order_payment'
# with columns: customer_city, order_id

# Group orders by city and calculate total order count
state_order_counts = merge_order_payment.groupby('customer_state')['order_id'].count()

# Plot a bar chart to show geographic trends
plt.figure(figsize=(10, 6))
plt.bar(state_order_counts.index, state_order_counts.values)
plt.xlabel('State')
plt.ylabel('Total Order Count')
plt.title('Purchase Behavior by State')
plt.xticks(rotation=45)
plt.show()
```
